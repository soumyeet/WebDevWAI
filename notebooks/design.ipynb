````python
import google.generativeai as genai
import yaml
import os

# Load API key
with open('config.yaml', 'r') as f:
    config = yaml.safe_load(f)
GEMINI_API_KEY = config['GEMINI_API_KEY']
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash-exp')

def save_file(filename: str, content: str) -> str:
    """
    Saves raw text in the content string to a file specified by filename string
    """
    try:
        # Ensure directory exists
        file_path = os.path.abspath(filename)
        
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(content)

        print(f"File saved successfully: {file_path}")
        return file_path

    except Exception as e:
        print(f"Error saving file: {e}")
        return None
````

````python
def response(prompt, name):
    chat = model.start_chat(enable_automatic_function_calling=True)
    response = chat.send_message(prompt)
    generated_code = response.text
    html_code = ""
    response_txt = "" 

    if "<html>" in generated_code:
        html_code = generated_code.split("<html>")[1].split("</html>")[0]
        html_code = f"<html>{html_code}</html>"  # Re-add the tags
    if "```" in generated_code:
        response_txt = generated_code.split("```")[1].split("```")[0]

    save_file(f"index_{name}.html", html_code)
    save_file(f"index_{name}.txt", response_txt)
````
